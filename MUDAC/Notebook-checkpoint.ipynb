```python
import pandas as pd
 
# Creating the dataframe 
df = pd.read_excel("C:/Users/user/Desktop/MUDAC/Training.xlsx")
 

df.info()
```

```python
from textblob import TextBlob

def get_sentiment(text):
    analysis = TextBlob(text)
    return analysis.sentiment.polarity  # Ranges from -1 (negative) to 1 (positive)

df['Match Support Contact Notes'] = df['Match Support Contact Notes'].astype(str)
df['Big_Sentiment_Score'] = df['Match Support Contact Notes'].apply(get_sentiment)


df['Closure Details'] = df['Closure Details'].astype(str)
df['Closure_Sentiment_Score'] = df['Closure Details'].apply(get_sentiment)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate correlation matrix
correlation_matrix = df.corr()

plt.figure(figsize=(13, 13))  # Adjust width and height as needed

# Create heatmap with adjustments to prevent overlapping
sns.heatmap(correlation_matrix, 
            annot=True,              # Show numbers
            cmap='coolwarm',         # Color scheme
            vmin=-1, vmax=1,         # Correlation range
            annot_kws={'size': 10},  # Reduce font size of annotations
            fmt='.3f',              # Format numbers to 3 decimal places
            square=True,            # Optional: makes cells square
            linewidths=0.5)         # Optional: adds spacing between cells

plt.title('Correlation Matrix: Sentiment Score vs Match Length', pad=20)
plt.tight_layout()  # Adjusts layout to prevent clipping
plt.show()
```
