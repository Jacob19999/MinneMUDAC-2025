```python
# Import required libraries
import geopandas as gpd

# Path to your shapefile (replace with your file path)
shp_file_path = "tl_2024_27_bg.shp"

# Step 1: Read the shapefile
gdf = gpd.read_file(shp_file_path)

# Step 2: Ensure the CRS is geographic (WGS84, EPSG:4326)
if gdf.crs != "epsg:4326":
    gdf = gdf.to_crs(epsg=4326)

# Step 3: Calculate centroids
gdf["centroid"] = gdf.geometry.centroid

# Step 4: Extract latitude and longitude
gdf["longitude"] = gdf["centroid"].x
gdf["latitude"] = gdf["centroid"].y

# Step 5: Select relevant columns (e.g., block group ID, lat, long)
result = gdf[["GEOID", "latitude", "longitude"]]  # Adjust 'GEOID' to your shapefile's ID column name

# Step 6: Display the first few rows
print(result.head())

# Optional: Save to CSV
result.to_csv("mn_block_group_centroids.csv", index=False)
print("Results saved to 'mn_block_group_centroids.csv'.")
```
